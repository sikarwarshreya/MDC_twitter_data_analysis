```python
search_results=api.search_tweets(q=('covid19'),lang='en',count=1000,result_type='mixed',include_entities=True,tweet_mode='extended')
print(search_results[0:2])
```

```python
rough=""
for tweets in search_results:
    if not tweets.retweeted:
        try:
            rough+=tweets.tweeted_status.full_text
        except AttributeError:
            rough+=tweets.full_text
    else:
        continue
```

```python
rough
```

```python
columns = set()
allowed_types = [str,int]
tweets_data = []
for status in search_results:
    status_dict = dict(vars(status))
    keys = status_dict.keys()
    single_tweet_data = {}
    for k in keys:
        try:
            v_type = type(status_dict[k])
        except:
            v_type = None
            
        if v_type != None:
            
            if v_type in allowed_types:
                single_tweet_data[k] = status_dict[k]   
                columns.add(k)
    tweets_data.append(single_tweet_data)
```
